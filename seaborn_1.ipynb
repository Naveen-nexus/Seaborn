```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# scatter plot
import seaborn as sns
tips = sns.load_dataset("tips")
tips
```

```python
# ṣcatter plot ->axis level function
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "sex", style="time", size="size")
```

```python
# jo ki mene scatter plot graph use kiya haio ya to ye figure level hoga ya asix levelhoga
# ye scatter plot graph axis level graph hota hai
# ye graph ham sam plot axes level plot ke sath bhi plot kar sakte ho
# uska naam hota hai relplot
```

```python
# relplot ->ek figur level function jo ki toda square shape me aata hai default hai
# konsa use karna hai
sns.relplot(data =tips, x= "total_bill" , y ="tip", kind="scatter", hue="sex", style ="time", size="size")
# kya ek ek dot customer present kar raha hai par mai chaata hu ki muje ek ek ki particular information chaiye ki vo male hai ya female
# hue parameter
# style parameter kya kar sakte hi ek aur layer  of information add kar sakte ho
# size parameter muje dekhna hai har customer kitne aur logo ke sath aya
```

```python
# seaborn bolta hai aap figure level function level function jayada use karo kyu ki aap ek figure level function ke use kaeke aap kai sare axes ko replicat kar sakte ho
```

```python
# line plot
# hota to scatter plot hi hota hai agar points ko jod do to vo line plot ho jata hai
# agar x->pe time hai kisi bhi form me years me months me days me to line plot use karte hai
```

```python
# line plot
import plotly.express as px
gap = px.data.gapminder()
temp_df = gap[gap["country"] =="India"]
# axes level function
sns.lineplot(data = temp_df, x = "year", y = "lifeExp")
# relplot
sns.relplot(data = temp_df, x = "year", y = "lifeExp", kind="line")
```

```python
# muje dekhna hai india ka lifeexpect 1952 se 2007 kese vary kiya
# first you fiter this data set on top of country column
```

```python
# hue ->style
temp_df = gap[gap["country"].isin(["India", "Brazil","Germany"])]
temp_df
```

```python
sns.relplot(kind="line", data=temp_df, x = "year", y = "lifeExp", hue="country",style ="continent")
```

```python
sns.lineplot(data=temp_df, x = "year", y = "lifeExp", hue="country",size="continent")
```

```python
# facet plot ->hota kya hai apke pass option hota hai facet plot use karne ka jiska use karke app ek particular catgorical column ke uper multiple plots bana sakte ho
# facet plot only work with figure level function  ->work with relplot
# it will not work with scatter level function and line plot function
sns.relplot(data=tips, x = "total_bill",y ="tip", kind="scatter", hue ="sex")
# aap hue parameter ki jagah col parmater use karoge
sns.relplot(data=tips, x = "total_bill",y ="tip", kind="scatter", col ="sex")
# column ki jagah rows bhi use kar sakte hai
sns.relplot(data=tips, x = "total_bill",y ="tip", kind="scatter", row ="sex")
# dono bhi use kar sakte hai
sns.relplot(data=tips, x = "total_bill",y ="tip", kind="scatter", col ="smoker",row ="sex")

```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# histplot
# histogram ->bins create kar rahe ho
```

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips
```

```python
# plotting univariant histogram

sns.histplot(data = tips, x = "total_bill")
```

```python
#using displot
sns.displot(data = tips, x = "total_bill", kind ="hist")
```

```python
# bins parameter
# for example muje bins badane hai -> hai to bins patle honge to bar bhi patl ho0o jayenege
```

```python
sns.displot(data=tips, x='total_bill', kind='hist',bins=2)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables
# generally ham histogram plot karte hai numerical data pe par categorical me bhi plot kar sakte hai
tips
```

```python
# countplot
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist', hue='sex')
# agar sirf outline banani hai to ek element diya hai element -"step"
sns.displot(data=tips, x='tip', kind='hist', hue='sex',element ="step")
```

```python
# defalut seaboen me aata hai datafram
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.displot(data =titanic, x= "age", hue="sex",kind = "hist", element ="step")
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step')
```

```python
# *************************kdeplot***************************
# kde = kernal density estimation
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
# same graph using displot
# hur paramter exist
sns.displot(data=tips,x='total_bill',kind='kde',hue ="sex",fill =True)
```

```python
# ***************************Rugplot*********************
# isko use karte hai agar ek plot bana liya uske support me bhaout use karte hai
# same distribution represent karta haib by plotting ticks on x and y axis alag se dikayi se jata hai
```

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
# jo orange me hai vo hai rug plot ye kya karta hai same information ko alaga form me dikhata hai
# sidewise hi use kiya jata hai aglag se  plot nahi kia jata
```

```python
# agar aap chaho to app bivariant histogram bhi bana sakte ho
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
# ek histogram bana jo ki 2D me hai aur jaha jaha intersect kiya unko app bol sakte ho chote chote grids aur un grids ka jinka jayada dark hai vo particular grids me fall kar rahe hai

# 2nd way dusra syntax
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# heatmap
#  it plots a rectangular data as a color-encoded matrix

# agar koi bhi data ko wide me convert karna hai to ham use karte hai pivot parameter
# pivot me aap batoage index kya hoga, columns kya hoga


# heatmap ek axes level function hota hai
```

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
```

```python
# annot ->number bhi show hone lag jayega
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
```

```python
# Clustermap ->very similar to headmap but isme kya hota hai ki ye internally aapke data ke uper clustering karta hai  similar coulumn ko ek sath le aata hai dissimilar do hata deta hai

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.
import plotly.express as px
iris = px.data.iris()
iris
```

```python
# mereko janna hai kaun se similar hai aur kaunse dissimilar hai
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
